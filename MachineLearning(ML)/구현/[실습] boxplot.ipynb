```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
body_data = pd.read_csv("/content/drive/MyDrive/코드잇/Pandas/수업/data/body.csv")
```

```python
q1 = np.percentile(body_data, 25)
q2 = np.percentile(body_data, 50)
q3 = np.percentile(body_data, 75)
iqr = q3 - q1

height_data = body_data['height']
weight_data = body_data['weight']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(height_data)
plt.title("키 박스")

plt.subplot(1, 2, 2)
sns.boxplot(weight_data)
plt.title("몸무게 박스")

plt.tight_layout()

plt.show()
```

# 두 변수 이차원 배열 넣어서 박스 플롯 그리기

```python
plt.boxplot([height_data, weight_data], labels=['Height', 'Weight'], vert=False)

plt.title("Boxplot - 2D array")
plt.tight_layout()
plt.show()
```
