```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(0, 2 * np.pi, 100)

height_data = body_data['height']
weight_data = body_data['weight']

plt.figure(figsize=(4, 15))

plt.subplot(3, 1, 1)
plt.hist(height_data, bins = 20, label = '키')
plt.title("키 그래프")
plt.xlabel("키")
plt.legend()

plt.subplot(3, 1, 2)
plt.title("몸무게 그래프")
plt.hist(weight_data, bins = 20, edgecolor='white', color = 'gray', label = '몸무게')
plt.xlabel("몸무게")
plt.legend()

plt.subplot(3, 1, 3)
plt.title("키 / 몸무게 산점도")
plt.scatter(height_data, weight_data, color = 'red', alpha = 0.3)
plt.xlabel("키")
plt.ylabel("몸무게")

plt.tight_layout()
plt.show()
```
