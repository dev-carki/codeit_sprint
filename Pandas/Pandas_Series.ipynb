# Series 객체
- 인덱스 + 값으로 이루어진 1차원 데이터

```python
import pandas as pd
```

# 데이터 생성 - 리스트

```python
data = [1, 2, 3]
series = pd.Series(data)
series
```

### index 추가

```python
data = [1, 2, 3]
series = pd.Series(data, index=['A', 'B', 'C'])
series
```

# 데이터 생성 - 딕셔너리

```python
data_dic = {
    'A': 1,
    'B': 2,
    'C': 3
}
series_dic = pd.Series(data_dic)
series_dic
```

# 실습

```python
kakao = pd.Series([92600, 92400, 92100, 94300, 92300])
kakao
```

### 인덱스를 통한 데이터 접근

```python
print(kakao[0])
print(kakao[3])
print(kakao[4])
```

### 리스트와 다른점
1. 인덱싱 값 지정 가능

```python
kakao2 = pd.Series(
    [92600, 92400, 92100, 94300, 92300],
    index=['2016-02-19', '2016-02-18', '2016-02-17', '2016-02-16', '2016-02-15']
)
kakao2
```

- 인덱스를 문자열로 지정했기 때문에, 데이터 접근도 해당 문자열로 접근해야됨

```python
print(kakao2['2016-02-15'])
```

# 인덱스와 값 출력

### 인덱스 출력

```python
for date_data in kakao2.index:
    print(date_data)
```

### 값 출력

```python
for price_data in kakao2.values:
    print(price_data)
```

```python
mine   = pd.Series([10, 20, 30], index=['naver', 'sk', 'kt'])
friend = pd.Series([10, 30, 20], index=['kt', 'naver', 'sk'])

mine + friend
```

### 인덱스가 다를 경우

```python
mine   = pd.Series([10, 20, 30], index=['naver', 'sk', 'kt'])
friend = pd.Series([10, 30, 20], index=['kt', 'naver', 'sk1'])

mine + friend
```
