```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x) # ÏÇ¨Ïù∏Ìï®Ïàò (ÏÇºÍ∞ÅÌï®Ïàò - ÏïîÍ∏∞ ÌïÑÏöî x)

plt.plot(x, y)
plt.title("ÏÇ¨Ïù∏ Í∑∏ÎûòÌîÑ") # -> ÌïúÍ∏Ä Íπ®Ïßê # ÌïúÍ∏ÄÌåêÏùÑ Îã§Ïö¥ Î∞õÏïÑÏÑú ÏÑ§ÏπòÌï¥ÏïºÎê®
plt.xlabel("X") # X Ï∂ïÏùò Ïù¥Î¶Ñ
plt.ylabel("Y") # Y Ï∂ïÏùò Ïù¥Î¶Ñ
plt.grid(True) # Í∑∏Î¶¨Îìú(ÎààÍ∏àÌëúÏãú)
plt.show()
```

# üü¢ Îã§Ï§ë Î†àÏù¥ÏïÑÏõÉ Í∏∞Î≥∏

<aside>

- Í∑∏ÎûòÌîÑ ÏÇ¨Ïù¥Ï¶à Ï°∞Ï†à
  - plt.figure(figsize=(10, 4))

- Í∑∏ÎûòÌîÑ Í≤πÏπ® Î∞©ÏßÄ
  - plt.tight_layout()
- Í∑∏ÎûòÌîÑ Î≤îÎ°Ä Ï∂îÍ∞Ä
  - plt.legend()

</aside>

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1) # 1Ìñâ (ÌïúÏ§Ñ) 2Ïó¥ (2Ïπ∏)Ïùò Ï≤´ Î≤àÏß∏ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≤†Îã§
plt.plot(x, np.sin(x), label="ÏÇ¨Ïù∏ Í∑∏ÎûòÌîÑ")
plt.title("ÏÇ¨Ïù∏(Sine) Í∑∏ÎûòÌîÑ")
plt.legend() # Í∑∏ÎûòÌîÑÏóê Î≤îÎ°Ä Ï∂îÍ∞Ä

plt.subplot(1, 2, 2) # 1Ìñâ (ÌïúÏ§Ñ) 2Ïó¥ (2Ïπ∏)Ïùò Îëê Î≤àÏß∏ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≤†Îã§
plt.title("ÏΩîÏÇ¨Ïù∏(Cosine) Í∑∏ÎûòÌîÑ")
plt.plot(x, np.cos(x), color='red', label="ÏΩîÏÇ¨Ïù∏ Í∑∏ÎûòÌîÑ")
plt.legend() # Í∑∏ÎûòÌîÑÏóê Î≤îÎ°Ä Ï∂îÍ∞Ä

plt.tight_layout() # Í∑∏ÎûòÌîÑ ÏÇ¨Ïù¥Ï¶à Ï°∞Ï†à / ÎààÍ∏àÎÅºÎ¶¨ Í≤πÏπòÎäîÍ±∞ Î∞©ÏßÄ
plt.show()
```

# üü¢ Îã§Ï§ë Î†àÏù¥ÏïÑÏõÉ Ïó¨Îü¨Í∞ú

```python
# import random
# import¬†matplotlib.colors¬†as¬†mcolors
#
# def¬†generate_random_color():
# ¬†¬†¬†¬†"""Generate¬†a¬†random¬†color¬†hex¬†string."""
# ¬†¬†¬†¬†colors¬†=¬†dict(mcolors.BASE_COLORS,¬†**mcolors.CSS4_COLORS)
# ¬†¬†¬†¬†color_names¬†=¬†list(colors.keys())
# ¬†¬†¬†¬†#¬†Remove¬†white¬†and¬†black¬†colors
# ¬†¬†¬†¬†color_names.remove('w')
# ¬†¬†¬†¬†color_names.remove('k')
# ¬†¬†¬†¬†#¬†Generate¬†a¬†random¬†color¬†name
# ¬†¬†¬†¬†color_name¬†=¬†random.choice(color_names)
# ¬†¬†¬†¬†#¬†Convert¬†the¬†color¬†name¬†to¬†hex
# ¬†¬†¬†¬†color_hex¬†=¬†mcolors.to_hex(colors[color_name])
# ¬†¬†¬†¬†return¬†color_hex

x = np.linspace(0, 2 * np.pi, 100)
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.plot(x, np.sin(x+i), color='red', label = f"ÏÇ¨Ïù∏(x+{i}) Í∑∏ÎûòÌîÑ")
    plt.title(f"Í∑∏ÎûòÌîÑ {i+1}")
    plt.xticks([]) # xÏ∂ï ÎààÍ∏à ÏóÜÏï†Í∏∞
    plt.yticks([]) # yÏ∂ï ÎààÍ∏à ÏóÜÏï†Í∏∞
    plt.legend() # Í∑∏ÎûòÌîÑÏóê Î≤îÎ°Ä Ï∂îÍ∞Ä

plt.tight_layout()
plt.show()
```
