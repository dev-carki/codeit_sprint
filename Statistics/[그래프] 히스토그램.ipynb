# KDE

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 드라이브 마운트 코드
from google.colab import drive

# Restart the runtime before running this cell.
drive.mount('/content/drive')
```

```python
body_data = pd.read_csv("/content/drive/MyDrive/코드잇/Pandas/수업/data/body.csv")
```

```python
body_data['height'].plot(kind='kde')
plt.grid(True)
plt.show()
```

# seaborn 이용
- 히스토그램 + KDE 그리기
- kde=True 넣으면 색

```python
sns.histplot(body_data['height'], color='gray')
plt.title("히스토그램+KDE")
plt.show()
```

# 선 색상 다르게
- stat

```python
# 1. 히스토그램 먼저 그리기
sns.histplot(body_data['height'], stat="density", kde=False)
sns.kdeplot(body_data['height'], color='red', linewidth=2)

plt.title("히스토그램+KDE 선 색 조정")
plt.show()
```

# 커넬폭 지정
- bw_method -> 기본값: 1
- 값이 적어질수록 왜곡 커짐

```python
body_data['height'].plot(kind='kde', bw_method=1)
plt.show()
```
